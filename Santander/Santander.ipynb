```python
"""Let's split the train and test set to feed machine learning algorithm."""
df_train = merged.iloc[:76020, :]
df_test  = merged.iloc[76020:, :]
```

```python
'''Drop passengerid from train set and Survived from test set.'''
df_train = df_train.drop(columns = ['ID'], axis = 1)
df_test = df_test.drop(columns = ['TARGET'], axis = 1)
```

```python
'''Extract data sets as input and output for machine learning models.'''
X_train = df_train.drop(columns = ['TARGET'], axis = 1) # Input matrix as pandas dataframe (dim:891*47).
y_train = df_train['TARGET'] #Output vector as pandas series (dim:891*1)
```

```python
"""Extract test set"""
X_test  = df_test.drop("ID", axis = 1).copy()

'''See the dimensions of input and output data set.'''
display(X_train.shape, X_test.shape, y_train.shape)
```

```python
from catboost import CatBoostClassifier
cbc = CatBoostClassifier()
```

```python
cbc.fit(X_train, y_train)
train_accuracy = cbc.score(X_train, y_train)
train_accuracy = np.round(train_accuracy*100, 2)
display(train_accuracy)
```

```python
from sklearn.model_selection import cross_val_score
x_val_score = cross_val_score(cbc, X_train, y_train, cv = 10, scoring = 'accuracy').mean()
x_val_score = np.round(x_val_score*100, 2)
```

```python
display(x_val_score)
```

```python
submission = pd.DataFrame({
        "PassengerId": test["PassengerId"],
        "Survived": cbc.predict(X_test)})
submission.to_csv('submission_rf.csv', index = False)
```
